```python
class MeanRegressors:
  def fit(self,inputs, targets):
    self.mean =np.mean(targets)
  
  def predict(self, inputs):
    return np.full(inputs.shape[0] , self.mean)
```

```python
mean_model=MeanRegressors()
```

```python
mean_model.fit(train_inputs,train_target)
```

```python
mean_model.mean
```

After the model is fitted, predicting the data by passing test_input dataset to the model using .predict()

```python
train_predsMean=mean_model.predict(train_inputs)
test_predsMean=mean_model.predict(test_inputs)
```

Evaluation metric used here is weighted mean absolute error (WMAE). Creating a function weight to calculate the weight w for a given dataset. Adding weight as 5 if it is a holiday else keeping the value as 1.

```python
def weight(df):
  if df['IsHoliday']==True:
    weight=5
  else:
    weight=1
  return weight 
```

```python
def WMAE(dataset, real, predicted,weights):
    return np.round(np.sum(weights*abs(real-predicted))/(np.sum(weights)), 2)
```

```python
WTR=train_inputs.apply(weight, axis=1)
WTE=test_inputs.apply(weight, axis=1)
```

```python
WTR.unique()
```

```python
WMAE(train_inputs,train_target,train_predsMean,WTR),WMAE(test_inputs,test_target,test_predsMean,WTE)
```

Observation is with the base model,the Weighted Mean Absolute Error for MeanRegressorModel predicts $15383.02 for train data and $15288.6 for Test Data.

Creating a new weight function based on holiday to calculate WMAE with all the added features

```python
def weightAdded(df):
  if df['IsHoliday_False']==1.0:
    weight=5
  else:
    weight=1
  return weight 
```

```python
def WMAE_updated(dataset, real, predicted,weights):
    return np.round(np.sum(weights*abs(real-predicted))/(np.sum(weights)), 2)
```

Implemented the logic to create the Weighted Mean Absolute Error.
Creating a function evaluate_updated to predict the Weekly Sales, then evaluate the WMAE of the train and test data Set.

```python
def evaluate_updated(model):
  train_preds=model.predict(train_inputs_added)
  check_train=train_inputs_added.apply(weightAdded,axis=1)
  train_wmae=WMAE_updated(train_inputs_added,train_target,train_preds,check_train)
  test_preds=model.predict(test_inputs_added)
  check_test=test_inputs_added.apply(weightAdded,axis=1)
  test_wmae=WMAE_updated(test_inputs_added,test_target,test_preds,check_test)
  return train_wmae,test_wmae,train_preds,test_preds
```

# HyperParameter Tuning

We will create a function test_params that accepts parameters and returns the WMAE for Training and Test set.

```python
def test_params( model,**params):
    """Trains a model with the given parameters and returns training & validation WMAE"""
    model=model(n_jobs=-1,random_state=42, **params)
    model.fit(train_inputs_imp,train_target)
    return evaluate_updated_imp(model),model
```

```python
def test_param_for_graph(**params):
    model=XGBRegressor(n_jobs=-1,random_state=42,**params)
    model.fit(train_inputs_imp,train_target)
    output=evaluate_updated_imp(model)
    train_wmae, test_wmae=output[0],output[1]
    return train_wmae, test_wmae
```

```python
# Function to plot results with different n_estimators

def test_param_plot(param_name, param_value):
  train_errors, test_errors=[],[]
  for i in param_value:
    params = {param_name: i}
    train_wmae, test_wmae = test_param_for_graph(**params)
    train_errors.append(train_wmae)
    test_errors.append(test_wmae)
  plt.figure(figsize=(16,8))
  plt.title('Overfitting curve: ' + param_name)
  plt.plot(param_value, train_errors, 'b-o')
  plt.plot(param_value, test_errors, 'r-o')
  plt.xlabel(param_name)
  plt.ylabel('WMAE')
  plt.legend(['Training', 'Test'])
```

```python
# Plotting with 200, 500 and 600 n_estimators

test_param_plot('n_estimators',[200,500,600])
```

From the above plot, we observe that the WMAE is minimum for n_estimator having values of 500 and 600

# Summary

Explored the Walmart data for 45 stores and predict sales based on the available historical data and identify whether factors like temperature, unemployment, fuel prices, CPI etc affect the weekly sales of these stores under study. The aim is to understand whether sales are relatively higher during holidays like Christmas and Thanksgiving, Labor day and Super Bowl than normal days so that stores can work on creating promotional offers that increase sales and generate higher revenue.
Walmart runs several promotional markdown sales throughout the year on days immediately following the prominent holidays in the United States which are taken into consideration to determine the impact on weekly sales and to drive resources towards such key strategic initiatives.

Extensive Exploratory data analysis is done for the Walmart dataset to understand the following: 

â€¢ Identifying store as well as department-wide sales in Walmart 
â€¢ Identifying sales based on store size and type 
â€¢ Identifying how much sales increase during holidays 
â€¢ Correlation between the different factors that affect sales 

Data Cleaning

Deletion of the unnamed columns that were not required for analysis. The records with weekly sales less than 0 were removed to avoid the error because of negative values. Temperature outliers were identified in the data but can be ignored. Markdown columns have the null values that were handled.

A new column was created as a markdown sum and imputed all the 5 markdown missing fields with the mean of the available column values. This is done to ensure the errors because of non-numeric values in such fields.

Feature Engineering

Considering four holiday weeks in the data, added the columns for super_bowl, Thanksgiving, Labor_Day and Christmas. The date field was split to get more insights about monthly, day-wise or yearly sales. Holiday_month column was introduced to explore the sales. The holiday months marked are February, September, November and December. Days left for Thanksgiving and Christmas are created with the expectation that sales will be higher before the holidays than on the exact holiday.

Numeric column scaling performed from range 0,1. Encoding of the categorical columns like Type and IsHoliday to numeric using OneHotEncoder.

The accuracy technique used is WMAE - Weighted Mean Absolute error.The weights are added based on the holiday flag, if it is a holiday week value is taken as 5 otherwise 1. n is the result dataset size that is the test dataset with 84042 records.

The data is split into test and train datasets based on date range. The ratio used is 80% for test and 20% for train data. Test data have records till date April 20th  2012 and Train data have records starting from April 20th 2012 till December 2012. First, the hardcoded model K-Mean regressor is trained. Further training the baseline model, Linear regression algorithm resulted in wmae as 14302.2

Created and trained various models to predict future sales using different algorithms. Gradient Booster seems to be the best algorithm for this data. It gives the minimum WMAE 2886 for the predicted values. The second-best model is the Random Forest Regression model.

Further analysed the feature importance using Random Forest regressor and Gradient Booster algorithms. Observation was theThanksgiving, Department, and Size have the maximum importance whereas MarkDown1, MarkDown2, MarkDown3, MarkDown4, MarkDown5, and Temperature have the least importance.
